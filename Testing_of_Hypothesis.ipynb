**Null Hypothesis (H0): There is no increase in sales after entering into Digital Marketing**

**Alternate Hypothesis (HA): There is increase in sales after stepping into Digital Marketing**

**Let Level of Significance = 0.05**

```python
from scipy.stats import ttest_ind
```

```python
t_test,p_value=ttest_ind(data['Sales_before_digital_add(in $)'],data['Sales_After_digital_add(in $)'])
```

```python
t_test
```

```python
p_value
```

```python
if p_value<0.05:
  print('Null Hypothesis is Rejected')
else:
  print('Failed to Reject Null Hypothesis')
```

Hence there is increase in sales since entering Digital Marketing
